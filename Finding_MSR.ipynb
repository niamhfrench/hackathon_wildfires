```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fwi_data = np.load('UK_FWI.npy')
```

```python
# Getting daily severity rating
fwi = fwi_data[4,:,:,:]
dsr = 0.0272*fwi**1.77
```

```python
msr = np.zeros((int(dsr.shape[0]/30),dsr.shape[1],dsr.shape[2]))
for i in range(int(dsr.shape[0]/30)):
    msr[i] = np.average(dsr[i*30:(i+1)*30,:,:],axis = 0)
```

```python
print(msr.shape)

plt.plot(msr[:,20,39])
```

```python
np.save("UK_MSR.npy", msr)
```
